# Node Classification with Node2Vec

# Introduction
This notebook will demonstrate the implementation of a pipeline that solves classification problems using graphs. The fundamental algorithm for the solution is Node2Vec.

#### What is Node2Vec?

A challenge when working with networks is transforming the network structure into a numerical representation that can be transferred to machine learning algorithms. Node2Vec is an algorithm that allows the user to map nodes from a graph G to an embedding. Generally, the embedding space has dimensions smaller than the number of nodes in the original graph G. The algorithm attempts to preserve the initial structure within the original graph. Essentially, similar nodes within the graph will generate similar embeddings in the embedding space. These embedding spaces are, in essence, a vector that corresponds to each node in the network. Graph embeddings are commonly used as input features to solve machine learning problems oriented towards link prediction, community detection, classification, etc.

#### What is Node Classification?
Node classification is a machine learning application to graphs. Typically, a classification model is trained to learn which class a node belongs to. This approach to both, binary and multiclass classification. Binary classification works with two different classes, while multiclass classification works with more than two. In this notebook, we will use Node2vec to generate node embeddings in the network. Node2vec is designed to preserve the initial structure within the original graph.

### How does it work?
* Random Walks:

Node2Vec performs random walks across the graph, as if exploring the neighbors of each node.

These walks generate sequences of nodes, similar to sentences in natural language.

#### Word2Vec:

These sequences are then treated as text and trained with Word2Vec, a simple neural network model that learns word representations (in this case, nodes).
Thus, nodes that appear in similar contexts will have similar vectors.

Key parameters:

* p and q: control the walk bias (local vs. global exploration).

* dimensions: number of embedding dimensions.

* window, min_count, batch_words: Word2Vec training parameters.

Node2Vec Parameters
1. p (Return parameter)
Controls the probability of returning to the previous node during the walk.

* High value: lower probability of returning → encourages exploration.

* Low value: higher probability of returning → encourages local depth.

2. q (In-out parameter)
Controls the probability of exploring new nodes.

* High value: wider walk (explores distant nodes).

* Low value: deeper walk (explores close neighbors).

Together, p and q allow balancing between BFS (local exploration) and DFS (global exploration).

Word2Vec Training Parameters

3. window

Context size around each node in the sequences.

Example: if window=2, the model learns to predict nodes that are 2 steps away in the walk.

4. min_count

Minimum frequency for a node to be included in the training.

If a node appears fewer times than min_count, it is discarded.

5. batch_words

Number of nodes processed per batch during training.
This affects performance and memory usage.

```python
import networkx as nx
import pandas as pd
import numpy as np
import arxiv

from arxiv import Client, SortCriterion
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, matthews_corrcoef, confusion_matrix, classification_report
from node2vec import Node2Vec as n2v
```

```python
from logging import error


# Lista de consultas
queries = ['automl', 'machinelearning', 'data', 'physics', 'mathematics', 'recommendation system', 'nlp', 'neural networks']

def search_arxiv(queries, max_results=100):
    """
    Search arXiv for a list of queries and return a DataFrame with metadata.

    Parameters:
        queries (List[str]): Keywords to search on arXiv.
        max_results (int): Max number of results per query (default: 100).

    Returns:
        pd.DataFrame: DataFrame with columns: title, date, article_id, url, main_topic, all_topics, authors, year.
    """
    client = Client()
    records = []

    for query in queries:
        try:
            search = arxiv.Search(
                query=query,
                max_results=max_results,
                sort_by=SortCriterion.SubmittedDate,
                sort_order=arxiv.SortOrder.Descending
            )
            results = client.results(search)
            for res in results:
                records.append({
                    'title': res.title,
                    'date': res.published,
                    'article_id': res.entry_id,
                    'url': res.pdf_url,
                    'main_topic': res.primary_category,
                    'all_topics': res.categories,
                    'authors': [author.name for author in res.authors]
                })
        except Exception as e:
            print(f"❌ Error with query '{query}': {e}")

    df = pd.DataFrame(records)
    if not df.empty:
        df['year'] = pd.DatetimeIndex(df['date']).year
        unique_ids = df.article_id.unique()
        id_map = {art: idx for idx, art in enumerate(unique_ids)}
        df['article_id'] = df['article_id'].map(id_map)
    return df

# Ejecutar búsqueda
research_df = search_arxiv(queries=queries, max_results=250)
print(f"🔍 Total papers found: {research_df.shape[0]}")
```

###Create a network
Now that we have obtained the data using the arXiv API, we can generate a network. The network will have the following structure: the nodes will be the article IDs (article_ids), and the edges will be all the topics that connect a pair of articles. For example, article ID 1, with the topics astrophysics and statistics, can be connected to article ID 10, with the topic statistics, and article ID 7, with the topics astrophysics and mathematics. This will be a multi-edge network where each edge will have a weight of 1.

```python
def generate_network(df, node_col, edge_col):
    '''
    This function will generate a article to article network given an input DataFrame.
    It will do so by creating an edge_dictionary where each key is going to be a node
    referenced by unique values in node_col and the values will be a list of other nodes
    connected to the key through the edge_col.

    params:
        df (DataFrame) : The dataset which holds the node and edge columns
        node_col (String) : The column name associated to the nodes of the network
        edge_col (String) : The column name associated to the edges of the network

    returns:
        A networkx graph corresponding to the input dataset

    example:
        generate_network(
            research_df,
            node_col = 'article_id',
            edge_col = 'main_topic'
        )
    '''
    edge_dct = {}
    for i,g in df.groupby(node_col):
        topics = g[edge_col].unique()
        edge_df = df[(df[node_col] != i) & (df[edge_col].isin(topics))]
        edges = list(edge_df[node_col].unique())
        edge_dct[i] = edges

    # create nx network
    g = nx.Graph(edge_dct, create_using = nx.MultiGraph)
    return g

tp_nx = generate_network(
    research_df,
    node_col = 'article_id',
    edge_col = 'main_topic'
)
```

```python
def graph_info(G):
    info = f"Type: {type(G).__name__}\n"
    info += f"Number of nodes: {G.number_of_nodes()}\n"
    info += f"Number of edges: {G.number_of_edges()}\n"
    info += f"Is directed: {G.is_directed()}\n"
    return info

print(graph_info(tp_nx))
```

###Apply Node2Vec
This component will cover running node2vec on the previously generated network and creating the associated node embeddings for that network. These embeddings will be crucial going forward, as they are the main features needed to build a node classification model.

* Dimensions: The node will be represented by a 16-dimensional embedding
The model is trained using parameters similar to those of Word2Vec:

* window: context size.
* min_count: minimum frequency to include a node.
* batch_words: training batch size.

The vectors are extracted from each node and a DataFrame is constructed where:
* Rows are nodes (articles).
* The columns are the dimensions of the embedding.

The embeddings DataFrame is joined with the articles DataFrame (research_df) using the article_id as an index. This allows each vector to be associated with its main topic (main_topic).

```python
g_emb = n2v(tp_nx, dimensions=16)

WINDOW = 1 # Node2Vec fit window
MIN_COUNT = 1 # Node2Vec min. count
BATCH_WORDS = 4 # Node2Vec batch words

mdl = g_emb.fit(
    window=WINDOW,
    min_count=MIN_COUNT,
    batch_words=BATCH_WORDS
)

emb_df = (
    pd.DataFrame(
        [mdl.wv.get_vector(str(n)) for n in tp_nx.nodes()],
        index = tp_nx.nodes
    )
)

emb_df = emb_df.merge(
    research_df[['article_id', 'main_topic']].set_index('article_id'),
    left_index = True,
    right_index = True
)

# Se imprime la frecuencia de los 10 tópicos más comunes en los artículos representados en el grafo.
print(emb_df.main_topic.value_counts().head(10))
```

### Train a classifier



```python
ft_cols = emb_df.drop(columns = ['main_topic']).columns.tolist()
target_col = 'main_topic'

# train test split
x = emb_df[ft_cols].values
y = emb_df[target_col].values

x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size = 0.3
)

# GBC classifier
clf = GradientBoostingClassifier()

# train the model
clf.fit(x_train, y_train)
```

### Evaluate the classifier

To evaluate the performance of this model, we will analyze the following accuracy indices:

* Accuracy Index
* Matthew's correlation coefficient
* Classification report
* Confusion matrix

```python
def clf_eval(clf, x_test, y_test):
    '''
    This function will evaluate a sk-learn multi-class classification model based on its
    x_test and y_test values

    params:
        clf (Model) : The model you wish to evaluate the performance of
        x_test (Array) : Result of the train test split
        y_test (Array) : Result of the train test split

    returns:
        This function will return the following evaluation metrics:
            - Accuracy Score
            - Matthews Correlation Coefficient
            - Classification Report
            - Confusion Matrix

    example:
        clf_eval(
            clf,
            x_test,
            y_test
        )
    '''
    y_pred = clf.predict(x_test)
    y_true = y_test

    y_pred = clf.predict(x_test)
    test_acc = accuracy_score(y_test, y_pred)
    print("Testing Accuracy : ", test_acc)

    print("MCC Score : ", matthews_corrcoef(y_true, y_pred))

    print("Classification Report : ")
    print(classification_report(y_test, clf.predict(x_test)))

    print(confusion_matrix(y_pred,y_test))

clf_eval(
    clf,
    x_test,
    y_test
)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import networkx as nx

from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report

# --- 1) Data preparation ---
# emb_df: DataFrame of embeddings (index = nodes), with 'main_topic' column after merge
# Secure feature columns
ft_cols = [c for c in emb_df.columns if c != 'main_topic']

# Only nodes that have tag to train
labeled = emb_df.dropna(subset=['main_topic']).copy()

X = labeled[ft_cols].values
y = labeled['main_topic'].values

# Reproducible and stratified split (provided there are >=2 per class)
x_train, x_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42)

# --- 2) Training ---
clf = GradientBoostingClassifier(random_state=42)
clf.fit(x_train, y_train)

print(classification_report(y_test, clf.predict(x_test)))

# --- 3) Prediction for ALL nodes in the graph ---
emb_df['pred_topic'] = clf.predict(emb_df[ft_cols].values)

# Confidence (maximum probability) to scale size
proba = clf.predict_proba(emb_df[ft_cols].values)
emb_df['pred_conf'] = proba.max(axis=1)
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import seaborn as sns
import pandas as pd
from matplotlib.patches import Patch

# -----------------------------------------------------------------
# Clean up duplicates from emb_df)
# -----------------------------------------------------------------
# This ensures the emb_df index is unique before plotting
if not emb_df.index.is_unique:
    print("⚠️ The emb_df index has duplicates. Aggregating by node...")

    # Define feature columns
    # Adjust this if you have other columns that aren't 'topic' or 'conf'
    feat_cols = [c for c in emb_df.columns if 'topic' not in c and 'conf' not in c]

    # Prepare the aggregation dictionary
    agg_dict = {**{c: 'mean' for c in feat_cols},
                'main_topic': (lambda s: s.mode().iat[0] if not s.mode().empty else s.iloc[0])}

    # Add predictions only if they exist (so as not to lose them)
    if 'pred_topic' in emb_df.columns:
        agg_dict['pred_topic'] = (lambda s: s.iloc[0])
    if 'pred_conf' in emb_df.columns:
        agg_dict['pred_conf'] = 'max'

    # Apply the aggregation
    emb_df = emb_df.groupby(level=0).agg(agg_dict)

    print("✅ emb_df cleaned and ready for plotting.")
else:
    print("ℹ️ emb_df already has a unique index. No cleaning needed.")


# -----------------------------------------------------------------
# PLOT THE REAL TOPIC (Ground Truth)
# -----------------------------------------------------------------

# 1) Align nodes (now there will be no mismatch)
nodelist = [n for n in tp_nx.nodes() if n in emb_df.index]
G_sub = tp_nx.subgraph(nodelist)

if len(nodelist) == 0:
    raise ValueError("There is no intersection between graph nodes and emb_df.")

# 2) Color palette by REAL TOPIC
topics = pd.Series(emb_df.loc[nodelist, 'main_topic']).astype(str).unique()
palette = sns.color_palette("tab20", n_colors=len(topics)) if len(topics) <= 20 else sns.color_palette("husl", n_colors=len(topics))
color_map = dict(zip(topics, palette))

# Node colors according to REAL TOPIC
node_colors = emb_df.loc[nodelist, 'main_topic'].astype(str).map(color_map).tolist()

# 3) Layout and drawing
pos = nx.spring_layout(G_sub, seed=42)

plt.figure(figsize=(12, 10))
nx.draw_networkx_edges(G_sub, pos, alpha=0.15, width=0.5)

# Now 'nodelist' and 'node_colors' will have the same length
nx.draw_networkx_nodes(
    G_sub, pos,
    nodelist=nodelist,
    node_color=node_colors,
    node_size=100, # Fixed size
    linewidths=0.6,
    edgecolors='white',
    alpha=0.95
)

# 4) Legend
topk_legend = 15
freq = emb_df.loc[nodelist, 'main_topic'].astype(str).value_counts()
top_topics = freq.head(topk_legend).index.tolist()
handles = [Patch(facecolor=color_map[t], label=f"{t} ({freq[t]})") for t in top_topics]
if len(freq) > topk_legend:
    handles.append(Patch(facecolor='lightgrey', label=f"... and {len(freq)-topk_legend} more"))

# Title and legend for REAL TOPIC
plt.legend(handles=handles, title='Real topic (frequency)', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.title("Graph colored by Real Topic (Ground Truth)")
plt.axis('off')
plt.tight_layout()
plt.show()
```
